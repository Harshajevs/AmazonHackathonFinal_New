```python
# Import necessary libraries
import torch
import torchaudio
import librosa
import numpy as np
from speechbrain.inference.interfaces import foreign_class
import warnings
warnings.filterwarnings('ignore')
```

```python
# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
# Load the pre-trained emotion recognition model
# This model achieves 78.7% accuracy on IEMOCAP dataset
classifier = foreign_class(
    source="speechbrain/emotion-recognition-wav2vec2-IEMOCAP",
    pymodule_file="custom_interface.py",
    classname="CustomEncoderWav2vec2Classifier"
)

print("SpeechBrain Wav2Vec2 emotion model loaded successfully!")
```

```python
def detect_audio_emotion_speechbrain(audio_path):
    """
    Detect emotion from audio file using SpeechBrain model

    Args:
        audio_path (str): Path to audio file

    Returns:
        dict: Dictionary containing emotion prediction and confidence
    """
    try:
        # Classify the audio file
        out_prob, score, index, text_lab = classifier.classify_file(audio_path)

        # Extract emotion label and confidence
        emotion = text_lab[0] if isinstance(text_lab, list) else text_lab
        confidence = float(score.max()) if hasattr(score, 'max') else float(score)

        # Map emotions to emojis
        emotion_emojis = {
            'neu': '😐 Neutral',
            'hap': '😊 Happy',
            'sad': '😢 Sad',
            'ang': '😠 Angry'
        }

        emotion_display = emotion_emojis.get(emotion, f"❓ {emotion}")

        return {
            'audio_file': audio_path,
            'emotion': emotion,
            'emotion_display': emotion_display,
            'confidence': confidence
        }

    except Exception as e:
        return {
            'error': f"Error processing audio: {str(e)}",
            'audio_file': audio_path
        }
```

```python

def display_emotion_result(result):
    """Display emotion detection results"""
    print("=" * 50)
    print("🎵 AUDIO EMOTION DETECTION RESULTS")
    print("=" * 50)

    if 'error' in result:
        print(f"❌ {result['error']}")
        return

    print(f"🎧 Audio File: {result['audio_file']}")
    print(f"🎯 Detected Emotion: {result['emotion_display']}")
    print(f"📊 Confidence Score: {result['confidence']:.4f}")
    print(f"📈 Confidence Percentage: {result['confidence']:.2%}")
    print("=" * 50)
```

```python
audio_file_path = "whoah.mp3"

# Detect emotion from audio
emotion_result = detect_audio_emotion_speechbrain(audio_file_path)
display_emotion_result(emotion_result)
```
